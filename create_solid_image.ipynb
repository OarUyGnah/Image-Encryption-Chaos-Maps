```python
#全黑的灰度图
gray0=np.zeros((500,500),dtype=np.uint8)
cv.imwrite('graphs\solid-color\\gray-black.png',gray0)

#全白的灰度图
gray0[:,:]=255
gray255=gray0[:,:]
cv.imwrite('graphs\solid-color\\gray-white.png',gray255)
#将灰度图转换成彩色图
Img_rgb=cv.cvtColor(gray255,cv.COLOR_GRAY2RGB)
#将RGB通道全部置成0
Img_rgb[:,:,0:3]=0
cv.imwrite('graphs\solid-color\\rgb-black.png',Img_rgb)
#将RGB通道全部置成255
Img_rgb[:,:,0:3]=255
cv.imwrite('graphs\solid-color\\rgb-white.png',Img_rgb)
```
